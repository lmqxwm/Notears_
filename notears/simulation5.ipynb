```python
semtypes = ["gauss", "gumbel", "uniform"]
fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(12, 12))
colormap = plt.cm.tab10
losses = ["loss_est", "loss_l1", "obj_aug", "obj_dual"]
loss_f = ["l", "F", "O", "D"]
#losses = ["loss_est_2", "loss_l1_2", "obj_aug_", "obj_dual_2"]
#losses = ["loss_est_t", "loss_l1_t", "obj_aug_t", "obj_dual_t"]
#losses = ["h", "h_2", "h_t"]
for i, sem_type in enumerate(semtypes):
    result = pd.read_csv("./results_loss/result_200_"+sem_type+".csv")
    ax = axs[i]
    for p, loss_ in enumerate(losses):
        ax.plot(result['d'], result[loss_], marker='o', label=loss_f[p], color=colormap(p))
    ax.set_xlabel('d')
    ax.set_ylabel("value")
    ax.set_title(f'sem_type={sem_type}, n=d-1')
    ax.legend()
plt.subplots_adjust(wspace=.2, hspace=.4)
plt.savefig("./results_loss/loss_nd_1.png")
plt.show()
```
