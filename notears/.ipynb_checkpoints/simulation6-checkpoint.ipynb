```python
import numpy as np
import utils
import linear
import pandas as pd
import scipy
import math
import scipy
import matplotlib.pyplot as plt
from itertools import permutations
import random
```

```python
utils.set_random_seed(1)
B_true = utils.simulate_dag(30, 30, "BP")
W_true = utils.simulate_parameter(B_true)
X = utils.simulate_linear_sem(W_true, 29, "gauss")
```

```python
B = 100
results = np.zeros(B+2)
results[0] = 0.5 / X.shape[0] * ((X-X@W_true) ** 2).sum()
all_perm = list(permutations(range(X.shape[1])))
perms = random.sample(all_perm, B)
r, _, _, _, _ = linear.notears_linear(X, W_true=W_true, lambda1=0, loss_type="l2")
results[1] = r
for i in range(B):
    print(f"permutations={i}")
    r, _, _, _, _ = linear.notears_linear(X[:, perms[i]], W_true=W_true, lambda1=0, loss_type="l2")
    results[i+2] = r
```

```python
plt.plot(results[1:])
```

```python
semtypes = ["gauss", "gumbel", "uniform"]
ds = [5, 10, 15, 20, 25, 30]
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
colormap = plt.cm.tab10
loss_f = ["loss_true", "loss_est", "loss_perm", "loss_max_perm"]
#losses = ["loss_est_2", "loss_l1_2", "obj_aug_", "obj_dual_2"]
#losses = ["loss_est_t", "loss_l1_t", "obj_aug_t", "obj_dual_t"]
#losses = ["h", "h_2", "h_t"]
for i, sem_type in enumerate(semtypes):
    result_d = np.zeros([len(ds), len(loss_f)+100])
    for j, d in enumerate(ds):
        result = pd.read_csv("./results_loss2/result_"+str(5000)+"_"+sem_type+"_"+str(d)+".csv")
        result_d[j, :2] = result.iloc[:2, 0]
        
        for k in range(100):
            result_d[j, k+2] = result.iloc[k+2, 0]

    ax = axs[i]
    for p in range(3):
        if p>0:
            ax.plot(ds, result_d[:,p], marker='o', label=loss_f[p], color=colormap(p))
    for p in range(100-1):
        ax.plot(ds, result_d[:,p+4], marker='', alpha=0.05, color=colormap(2))
    ax.set_xlabel('d')
    ax.set_ylabel("value")
    ax.set_title(f'sem_type={sem_type}, n=d-1, L-BFGS')
    ax.legend()
plt.subplots_adjust(wspace=.2, hspace=.4)
#plt.savefig("./results_loss/loss_nd_1.png")
plt.show()
```

```python
semtypes = ["gauss", "gumbel", "uniform"]
ds = [5, 10, 15, 20, 25, 30]
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
colormap = plt.cm.tab10
loss_f = ["loss_true", "loss_est", "loss_min_perm", "loss_max_perm"]
#losses = ["loss_est_2", "loss_l1_2", "obj_aug_", "obj_dual_2"]
#losses = ["loss_est_t", "loss_l1_t", "obj_aug_t", "obj_dual_t"]
#losses = ["h", "h_2", "h_t"]
for i, sem_type in enumerate(semtypes):
    result_d = np.zeros([len(ds), len(loss_f)+100])
    for j, d in enumerate(ds):
        result = pd.read_csv("./results_loss2/result_"+str(5000)+"_"+sem_type+"_"+str(d)+".csv")
        result_d[j, :2] = result.iloc[:2, 0]
        min_ind = np.argmin(result.iloc[2:, 0]) + 2
        result_d[j, 2] = result.iloc[min_ind, 0]
        max_ind = np.argmax(result.iloc[2:, 0]) + 2
        result_d[j, 3] = result.iloc[max_ind, 0]
        for k in range(100):
            result_d[j, k+4] = result.iloc[k+2, 0]

    ax = axs[i]
    for p in range(4):
        if p >-1:
            ax.plot(ds, result_d[:,p], marker='', label=loss_f[p], color=colormap(p))
    for p in range(100):
        ax.plot(ds, result_d[:,p+4], marker='', alpha=0.005, color=colormap(5))
    ax.set_xlabel('d')
    ax.set_ylabel("value")
    ax.set_title(f'sem_type={sem_type}, n=d-1, L-BFGS')
    ax.legend()
plt.subplots_adjust(wspace=.2, hspace=.4)
plt.savefig("./results_loss2/loss_perm_1.png")
plt.show()
```
